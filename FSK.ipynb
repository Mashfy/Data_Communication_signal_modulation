### **Frequency Shift Keying**

```python
v = input('Enter digital data:- ')
v = list(map(int, v))

dim=100
Vx=[]
Di=[]

for i in range (0,len(v)):
    f=np.ones(dim)
    x=f*v[i]
    Vx=np.concatenate((Vx,x))


fig = plt.figure(figsize = (10, 10))
fig.suptitle('Frequency Shift Keying', y=1.05, fontsize=18)

plt.subplot(4,1,1)
plt.plot(Vx, 'g')
plt.title("Message Signal")
plt.ylabel('Xm(t)')
dim2=len(Vx)

t=np.linspace(0,5,dim2)
plt.subplot(4,1,2)
w1=2*np.pi*5*t
y1=np.cos(w1)
plt.plot(t,y1, 'b')

plt.title("1st Carrier Signal")
plt.ylabel('Fc1(t)')
plt.subplot(4,1,3)
y2=np.cos(2*np.pi*20*t)
plt.plot(t,y2, 'b')

plt.title("2nd Carrier Signal")
plt.ylabel('Fc2(t)')
plt.subplot(4,1,4)

for i in range(0,dim2):
    if Vx[i]==0:
        cero=np.array([y1[i]])
        Di=np.concatenate((Di,cero))
    else:
        uno=np.array([y2[i]])
        Di=np.concatenate((Di,uno))
plt.plot(t,Di, 'k')
plt.title("FSK Modulated Signal")
plt.ylabel('Ym(t)')

fig.tight_layout(pad=1.0)
plt.show()
```
