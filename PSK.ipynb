### **Phase Shift Keying**

```python
v = input('Enter digital data:- ')
v = list(map(int, v))

dim=100
Vx=[]
Di=[]
#Ciclo para crear la trama de datos digitales
for i in range (1,len(v)):
    f=np.ones(dim)
    x=f*v[i]
    Vx=np.concatenate((Vx,x))


fig = plt.figure(figsize = (10, 10))
fig.suptitle('Phase Shift Keying', y=1.05, fontsize=18)

plt.subplot(4,1,1)
plt.plot(Vx, 'g')
plt.title("Message signal")
plt.ylabel('Xm(t)')
dim2=len(Vx)

t=np.linspace(0,5,dim2)
plt.subplot(4,1,2)
y1=np.cos(2*np.pi*3*t)
plt.plot(t,y1, 'b')

plt.title("Phase 0 Carrier Signal")
plt.ylabel('Fc1(t)')
plt.subplot(4,1,3)
y2=np.sin(2*np.pi*3*t)
plt.plot(t,y2, 'b')

plt.title("Phase 180 Carrier Signal")
plt.ylabel('Fc2(t)')
plt.subplot(4,1,4)
res=((y2*Vx)-(y1*Vx)+(y1))
plt.plot(t,res, 'k')

plt.title("PSK Modulated Signal")
plt.ylabel('Ym(t)')

fig.tight_layout(pad=1.0)
plt.show()
```
