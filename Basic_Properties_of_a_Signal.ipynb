### **Basic Properties of a Signal.**

#### **Reflection:** Let's consider a signal sin(t) ; the reflected signal is described by sin(-t). Thus, the reflected signal assumes at time â€“t the value of the original signal that occurs at time t. 

```python
#@title Reflection Code
t = np.arange(-(4*np.pi),4*np.pi,0.1)

p = np.sin(t)
q = np.sin(-t)


fig = plt.figure(figsize = (12, 6))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.plot(t,p, 'b-', label='y=sin(t)')
plt.plot(t,q, 'c-', label='y=sin(-t))')


plt.legend(loc='upper left')
plt.show()
```

### **Shifting:** 
####Let   **y=$\frac{1}{âˆš(2Ï€)}$e<sup>-t<sup>2</sup>/2</sup>**    be an original signal. To obtain the shifted signal, for a shift t<sub>0</sub>=2, the value of the original signal that took place at t must now occur at t+t<sub>0</sub> . If t<sub>0</sub> is positive, the shifted signal is called a delayed signal.


```python
#@title Shifting Code
t = np.linspace(-4, 4, 100)
p = 1/(2 * np.pi)**0.5 * np.exp(-t**2 / 2)
q = 1/(2 * np.pi)**0.5 * np.exp(-(t-2)**2 / 2)


# setting the axes at the centre
fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# plot the functions, with labels
plt.plot(t,p, 'b-', label='not shifted')
plt.plot(t,q, 'c-', label='shifted')


plt.legend(loc='upper left')

# show the plot
plt.show()
```

### **Periodicity:**
###A continuous signal is said to be periodic if there exists such a time interval T that x(t+T)= x(t); for all t. The smallest time T is known as the period. It should be noted that if a signal is periodic with the period T, it is also periodic for any integer multiple of T. In below, the Periodicity of sine fuction is shown.

```python
#@title Periodicity Code
t = np.arange(-(4*np.pi),4*np.pi,0.1)
p = np.sin(2*(np.pi)+t)
q = np.sin(t)


fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(t,p, 'b-', linewidth=4, label='x(t+T)')
plt.plot(t,q, 'c-', linewidth=2, label='x(t)')


plt.legend(loc='upper left')

plt.show()
```
