###Line Coding Functions

```python

def UnipolarNRZ(bit_seq):
    t=[]
    amp=[]
    x=[]
    dt = {'1':5, '0':0}

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
    t.append(i+1)
    x.append(i+1)

    for i in range(0, len(bit_seq)):
        amp.append(dt[bit_seq[i]])
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b')
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([0, 5])
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()

def UnipolarRZ(bit_seq):
    t=[]
    amp=[]
    x=[]

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
        if bit_seq[i]=='1':
            t.append(i+0.5)
    t.append(i+1)
    x.append(i+1)

    for i in range(0, len(bit_seq)):
        if bit_seq[i]=='0':
            amp.append(0)
        else:
            amp.append(5)
            amp.append(0)
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([0, 5])
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()
    
def PolarNRZL(bit_seq):
	t=[]
	amp=[]
	x=[]
	dt = {'1':5, '0':-5}

	for i in range(0, len(bit_seq)):
		x.append(i)
		t.append(i)
		
	t.append(i+1)
	x.append(i+1)

	for i in range(0, len(bit_seq)):
	    amp.append(dt[bit_seq[i]])
	amp.append(0)

	plt.figure()
	plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
	plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
	plt.xticks(x)
	plt.yticks([0, 5, -5])
	plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
	plt.show()

def PolarNRZI(bit_seq):
    t=[]
    amp=[]
    x=[]

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
    t.append(i+1)
    x.append(i+1)
    
    prev = -5
    
    for i in range(0, len(bit_seq)):
        if  bit_seq[i]=='1':
            if prev==5:
                amp.append(-5)
                prev=-5
            else:
                amp.append(5)
                prev=5
        else:
            amp.append(prev)
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([0, 5, -5])
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()

def PolarRZ(bit_seq):
    t=[]
    amp=[]
    x=[]
    dt = {'1':5, '0':-5}

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
        t.append(i+0.5)
    t.append(i+1)
    x.append(i+1)

    for i in range(0, len(bit_seq)):
        amp.append(dt[bit_seq[i]])
        amp.append(0)
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([0, 5, -5])
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()

def Manchester(bit_seq):
    t=[]
    amp=[]
    x=[]

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
        t.append(i+0.5)
    t.append(i+1)
    x.append(i+1)

    for i in range(0, len(bit_seq)):
        if bit_seq[i]=='0':
            amp.append(-5)
            amp.append(5)
        else:
            amp.append(5)
            amp.append(-5)
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([0, 5, -5])
    plt.tight_layout(pad=3.0)
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()

def DifferentialManchester(bit_seq):
    t=[]
    amp=[]
    x=[]

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
        t.append(i+0.5)
    t.append(i+1)
    x.append(i+1)
    
    prev=5
    for i in range(0, len(bit_seq)):
        if bit_seq[i]=='0':
            if prev==5:
                amp.append(-5)
                amp.append(5)
                prev=5
            else:
                amp.append(5)
                amp.append(-5)
                prev=-5
        else:
            if prev==5:
                amp.append(5)
                amp.append(-5)
                prev=-5
            else:
                amp.append(-5)
                amp.append(5)
                prev=5
                
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([0, 5, -5])
    plt.tight_layout(pad=3.0)
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()

def BipolarAMI(bit_seq):
    t=[]
    amp=[]
    x=[]

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
    t.append(i+1)
    x.append(i+1)
    
    prev=-5
    for i in range(0, len(bit_seq)):
        if bit_seq[i]=='1':
            if prev==5:
                amp.append(-5)
                prev=-5
            else:
                amp.append(5)
                prev=5
        else:
            amp.append(0)
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([-5,0, 5])
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()

def BipolarPseudoternary(bit_seq):
    t=[]
    amp=[]
    x=[]

    for i in range(0, len(bit_seq)):
        t.append(i)
        x.append(i)
    t.append(i+1)
    x.append(i+1)
    
    prev=-5
    for i in range(0, len(bit_seq)):
        if bit_seq[i]=='0':
            if prev==5:
                amp.append(-5)
                prev=-5
            else:
                amp.append(5)
                prev=5
        else:
            amp.append(0)
    amp.append(0)

    plt.figure()
    plt.hlines(0, t[0], t[len(t)-1]+1, colors='b', linewidth=5.0)
    plt.plot(t, amp, drawstyle='steps-post', linewidth=5.0, color='k')
    plt.xticks(x)
    plt.yticks([-5,0, 5])
    plt.grid(b=True, color='g', linestyle='--', linewidth=2.0, axis='x')
    plt.show()
```

###Encoding and Encoding Functions for 4B/5B Block Coding

```python

def get_Mapping_Codes(bits):
    return block_coding_mapping_codes["".join(map(str, bits))]
  
'''
bits = sequence of bits to encoded using mB/nB
'''
def get_MB_NB_Encoding(bits, mB, nB):
    # If the length of bits is not multiple of mB, appends extra 0's at the end of the bits(optional)
    if len(bits) % mB != 0:
      # Necessary bit count to make it multiple of mB
      extra_len = (((len(bits)// mB) + 1) * mB) - len(bits)
      extra = np.array(np.zeros(extra_len), dtype=int)
      bits = np.append(bits, extra)
    
    encoded_bits_str = ""
    for i in range(0, len(bits), mB):
      encoded_bits_str += get_Mapping_Codes(bits[i:i+mB])
    
    # Convert the encoded bit string to list of bits
    encoded_bits_list = list(encoded_bits_str)
    encoded_bits_map = map(int, encoded_bits_list)
    encoded_bits = np.array(list(encoded_bits_map))
    return encoded_bits

'''
bits = sequence of bits to decoded using mB/nB
'''
def get_NB_MB_Decoding(bits, mB, nB):
    if len(bits) % nB != 0:
      print("Invalid length, Decoding Not possible!")
      return
    decoded_bits_str = ""
    for i in range(0, len(bits), nB):
      decoded_bits_str += get_Mapping_Codes(bits[i:i+nB])
    # Convert the decoded bit string to list of bits
    decoded_bits_list = list(decoded_bits_str)
    decoded_bits_map = map(int, decoded_bits_list)
    decoded_bits = np.array(list(decoded_bits_map))
    return decoded_bits

def PolarNRZIDecoding(bits):
    Time = 10e-6   
    bit_rate = 1/Time  
    Fs = 4*bit_rate    
    N = len(bits)   
    T = np.linspace(0,Time, mB * 100)

    x = np.array([])
    last_bit = 1
    for i in range(N):
      if bits[i] == 1:
        temp = (-1) * last_bit * np.ones(len(T))
        last_bit = -last_bit
      else:
        temp = last_bit * np.ones(len(T))
      x = np.append(x, temp)
    t = np.linspace(0,N*Time,len(x))
    result = []
    counter = 0
    last_bit = 1
    for i in range(len(t)):
      if t[i]*bit_rate > counter:
        counter = counter + 1
        if x[i] != last_bit:
          result.append(1)
          last_bit=-last_bit
        else:
          result.append(0)
    return result

def Encodingwith4b5b(bit_seq):
    bits = list(map(int, bit_seq))
    bits = get_MB_NB_Encoding(bits, mB, nB)

    str1 = ''.join(str(e) for e in bits)
    PolarNRZI(str1) 

    loop=True  
    while loop:
        print("Do you want to decode now?:- ")
        print("1. 4B/5B Decoding")
        print("2. Exit")

        option=int(input("Enter your choice>>> "))

        if option==1:
            Decodingwith4b5b(bit_seq, mB, nB)
            loop=False  


def Decodingwith4b5b(bit_seq, mB, nB):

    bits=list(map(int, bit_seq.split(" ")))
    bits = list(map(int, bit_seq))
    bits = get_MB_NB_Encoding(bits, mB, nB)

    result = PolarNRZIDecoding(bits)
    print('Decoded by NRZ-I line coding:', result)

    decoded_result = get_NB_MB_Decoding(result, mB, nB)
    print('4B/5B Decoded result', decoded_result)
```
