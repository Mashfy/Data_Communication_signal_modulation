#### **Frequency Modulation**

```python
f_m = float(input('Enter modulator frequency: '))
f_c = float(input('Enter carrier frequency: '))
modulation_index = float(input('Enter modulation index: '))

t = np.linspace(0, 1, 1000)
modulator = np.sin(2.0 * np.pi * f_m * t) * modulation_index
carrier = np.sin(2.0 * np.pi * f_c * t)
product = np.zeros_like(modulator)

for i, t in enumerate(t):
    product[i] = np.sin(2. * np.pi * (f_c * t + modulator[i]))

fig = plt.figure(figsize = (10, 8))
fig.suptitle('Frequency Modulation', y=1.03, fontsize=18)

plt.subplot(3, 1, 1)
plt.title('Modulator signal')
plt.plot(modulator,'g')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.title('Carrier signal')
plt.plot(carrier,'b')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.title('FM signal')
plt.plot(product,'k')
plt.ylabel('Amplitude')


fig.tight_layout(pad=2)
plt.show()
```
