#### **Amplitude Shift Keying**



```python

v = input('Enter digital data:- ')
v = list(map(int, v))

dim=100
Vx=[]
for i in range (0,len(v)):
    f=np.ones(dim)
    x=f*v[i]
    Vx=np.concatenate((Vx,x))

fig = plt.figure(figsize = (10, 8))
fig.suptitle('Amplitude Shift Keying', y=1.05, fontsize=18)

plt.subplot(3,1,1)
plt.plot(Vx,'g')
plt.title("Message Signal")
plt.ylabel('Xm(t)')
dim2=len(Vx)

t=np.linspace(0,5,dim2)
plt.subplot(3,1,2)
y1=np.cos(2*np.pi*10*t)
plt.plot(t,y1,'b')
plt.title("Carrier Signal")
plt.ylabel('Fc(t)')

plt.subplot(3,1,3)
mult=(Vx*y1)
plt.plot(t,mult,'k')
plt.title("ASK Modulated Signal")
plt.ylabel('Ym(t)')

fig.tight_layout(pad=2.0)
plt.show()
```
